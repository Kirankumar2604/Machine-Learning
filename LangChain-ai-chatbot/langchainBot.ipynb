```python
from langchain.chat_models import init_chat_model
from langchain_core.prompts import PromptTemplate
import gradio as gr
```

```python
prompt_template_str = """
Your task is to explain the concept of **{concept}** to me in a way that is:

1. Clear and intuitive
2. Concise (under 100 words)
3. Tailored specifically to my background and learning goals

Personalize your explanation using the following information:

The personalization should be subtle and natural. Reference my background only when it genuinely improves the explanation. Avoid unnecessary or forced mentions.
"""
```

```python
os.environ["OPENAI_BASE_URL"] = "https://openrouter.ai/api/v1"

model = init_chat_model(
    "openai/gpt-4.1-mini",
    model_provider="openai",
    max_tokens=300,
)
```

```python
concept = "aiml"

prompt_template = PromptTemplate.from_template(prompt_template_str)
```

```python
def generate_text(input_text):
  # Format the prompt with the input variable
  prompt = prompt_template.format(concept=input_text)
  # Call the model with the prompt
  response = model.invoke(prompt)
  return response.text
```

```python
demo = gr.Interface(
    fn=generate_text,
    inputs=[gr.Textbox(label="Concept", lines=1)],
    outputs=[gr.Textbox(label="Explanation", lines=5)],
    flagging_mode="never",
    title="Concept Explainer",
    description="Enter any concept and get a personalized explanation"
)
demo.launch(share=True, debug=True)
```
