```python
def cal_space(x):
    value = x.split('-')
    if len(value) == 2:
        return (float(value[0])+float(value[1]))/2
    try:
        return float(x)
    except:
        return None    
```

```python

df4['cost_per_sqft'] = (df4['price'] * 1000000) / df4['total_sqft']

df4.head()
```

```python
df4.location.unique()
```

```python
df4.location = df4.location.apply(lambda x: x.strip())
```

```python
location_stats = df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats.head(30)
```

```python
len(location_stats[location_stats<10])
```

```python
location_stats_less_that_10 = location_stats[location_stats<10]
location_stats_less_that_10
```

```python
df4.location = df4.location.apply(lambda x: 'other' if x in location_stats_less_that_10 else x)
len(df4.location.unique())
```
